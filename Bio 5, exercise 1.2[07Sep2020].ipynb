```python
# algorithm to copy mirror elements from upper diagonal half to lower

for y in range(0, n-1):
    for x in range((y+1), n):
        dij[x][y] = dij[y][x] # copy element across diagonal 
    
print(dij)
```

```python
def ParseGraph(adjList): 
# parse adjList into starts and ends 
    starts = [i.split('->')[0] for i in adjList] # works good 
    ends  =  [m.split(':')[0] for m in [i.split('->')[1] for i in adjList]] # splits twice 
    edges =  [m.split(':')[1] for m in [i.split('->')[1] for i in adjList]] # splits twice 
#     output = numpy.matrix( [starts, ends, edges] ).transpose()
#     output =  [starts, ends, edges] 
    output = numpy.array( [starts, ends, edges] ).transpose()


    return output # returns 3 arrays corresponding to start nodes, end nodes, edge weights
```

```python
def DFSdistance(adjArray, start, end, visited, totLength):
    # find subset of adjList where adjList.starts = start, indicates nodes one step from start 
    print('\nstart, end, visited:', start, end, visited) # debug only
#     print('adjArray:', adjArray) # debug only
    
    if len(visited) >10: return "got stuck in loop" # safety for debug only
    
    # select nearEdges, edges that connect to current edge
    nearEdges = [list(i) for i in adjArray if i[0] == start ]
    nearEdges = numpy.array(nearEdges)
    print('nearEdges:', nearEdges) # debug only
    
    if nearEdges.size == 0:
        print('nearEdges empty, ended now', nearEdges) # debug only
        return None 
    
    for n in nearEdges: # for each node one-step-away
        print('n[1]', n[1], str(n[1]))
        if n[1] == end: # check if it's the final node, then search completed 
            totLength += int( n[2]) # add node length to total length
            print('found last node:', n[1], '+ length:', n[2], ", totLength",totLength) # debug only
            return [visited, totLength ] 
        if str(n[1]) not in visited: # check if edge.ends has NOT been visited
            visited.append(n[0]) # append node to visited list
            totLength += int( n[2]) # add node length to total length
            print('checked node:', n[1], '+ length:', n[2], ", totLength",totLength) # debug only
            # otherwise run again, designate node.end as new start node
            DFSdistance(adjArray, n[1], end, visited, totLength)  
            # need to add ONLY edges that contribute to path, not every edge queried
```
